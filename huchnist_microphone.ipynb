```python
# Ð’ Ð Ð•Ð–Ð˜ÐœÐ† Ð Ð•ÐÐ›Ð¬ÐÐžÐ“Ðž Ð§ÐÐ¡Ð£
import pyaudio
import os
import struct
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft
import time
from tkinter import TclError
import math

%matplotlib notebook
#inline

#20*LG[1Pa/20*10^(-6)Pa]=94 dB

#Ñ€Ñ–Ð²ÐµÐ½ÑŒ Ð·Ð²ÑƒÐºÐ¾Ð²Ð¾Ð³Ð¾ Ñ‚Ð¸ÑÐºÑƒ (SPL) Ð²Ñ…Ñ–Ð´Ð½Ð¾Ð³Ð¾ ÑÐ¸Ð³Ð½Ð°Ð»Ñƒ, dB
#Ñ€Ñ–Ð²ÐµÐ½ÑŒ Ð²Ð¸Ñ…Ñ–Ð´Ð½Ð¾Ð³Ð¾ ÑÐ¸Ð³Ð½Ð°Ð»Ñƒ, dBV

#Ñ‡ÑƒÑ‚Ð»Ð¸Ð²Ñ–ÑÑ‚ÑŒ Ð¼Ñ–ÐºÑ€Ð¾Ñ„Ð¾Ð½Ñƒ - Ñ†Ðµ Ð½Ð°Ð¿Ñ€ÑƒÐ³Ð°, ÑÐºÐ° Ð²Ð¸Ð½Ð¸ÐºÐ°Ñ” Ð½Ð° Ð²Ð¸Ñ…Ð¾Ð´Ñ– Ð¼Ñ–ÐºÑ€Ð¾Ñ„Ð¾Ð½Ð°*, 
#ÐºÐ¾Ð»Ð¸ Ð½Ð° Ð½ÑŒÐ¾Ð³Ð¾ Ð¿Ð¾Ð´Ð°Ñ‚Ð¸ ÑÐ¸Ð½ÑƒÑÐ¾Ñ—Ð´Ð°Ð»ÑŒÐ½Ð¸Ð¹ ÑÐ¸Ð³Ð½Ð°Ð» Ñ‡Ð°ÑÑ‚Ð¾Ñ‚Ð¾ÑŽ 1ÐºÐ“Ñ† Ñ– Ñ€Ñ–Ð²Ð½ÐµÐ¼ Ð·Ð²ÑƒÐºÐ¾Ð²Ð¾Ð³Ð¾ Ñ‚Ð¸ÑÐºÑƒ 94Ð´Ð‘**, 
#Ñ‰Ð¾ Ð²Ñ–Ð´Ð¿Ð¾Ð²Ñ–Ð´Ð°Ñ” Ñ‚Ð¸ÑÐºÑƒ 1ÐŸÐ°

#Ð‘Ñ–Ð»ÑŒÑˆÑ–ÑÑ‚ÑŒ Ð°Ð½Ð°Ð»Ð¾Ð³Ð¾Ð²Ð¸Ñ… ÐµÐ»ÐµÐºÑ‚Ñ€ÐµÑ‚Ð½Ð¸Ñ… Ð¼Ñ–ÐºÑ€Ð¾Ñ„Ð¾Ð½Ñ–Ð² Ð¼Ð°ÑŽÑ‚ÑŒ Ñ‡ÑƒÑ‚Ð»Ð¸Ð²Ñ–ÑÑ‚ÑŒ Ñƒ Ð´Ñ–Ð°Ð¿Ð°Ð·Ð¾Ð½Ñ– Ð²Ñ–Ð´ â€“46 Ð´Ð¾ â€“35 Ð´Ð‘Â·Ð’, 
#Ñ‚Ð¾Ð¼Ñƒ Ð±ÑƒÐ´ÐµÐ¼Ð¾ Ð²Ð²Ð°Ð¶Ð°Ñ‚Ð¸, Ñ‰Ð¾ Ñ‡ÑƒÑ‚Ð»Ð¸Ð²Ñ–ÑÑ‚ÑŒ Ð¼Ñ–ÐºÑ€Ð¾Ñ„Ð¾Ð½Ð° -40 Ð´Ð‘Â·Ð’   (-47)

#*Ð¼Ñ–ÐºÑ€Ð¾Ñ„Ð¾Ð½ Ð½Ð° Ð²Ñ–Ð´ÑÑ‚Ð°Ð½Ñ– 1 Ð¼ÐµÑ‚Ñ€ Ð²Ñ–Ð´ Ð³ÑƒÑ‡Ð½Ð¾Ð¼Ð¾Ð²Ñ†Ñ
#**Ñ€Ð¾Ð·Ð¼Ð¾Ð²Ð°, Ð½Ð° Ð²Ñ–Ð´ÑÑ‚Ð°Ð½Ñ– Ð±Ð»Ð¸Ð·ÑŒÐºÐ¾ 3 ÑÐ¼ Ð¿Ñ€Ð¸Ð±Ð»Ð¸Ð·Ð½Ð¾ Ð´Ð¾Ñ€Ñ–Ð²Ð½ÑŽÑ” Ð³ÑƒÑ‡Ð½Ð¾ÑÑ‚Ñ– 94 dB
 
#120 dB (SPL) ---  (-21) dBV ---20       Pa ---89.13 mV/Pa --- 32768=2^15
# 94 dB (SPL) ---  (-47) dBV --- 1       Pa --- 4.47 mV/Pa ---  1642.29
#  0 dB (SPL) --- (-141) dBV --- 0.00002 Pa --- 0    mV/Pa ---     0

def RA(v):
    g_1=12194**2*v**4/((v**2+20.6**2)*(v**2+12194**2)*((v**2+107.7**2)*(v**2+737.9**2))**0.5)
    v=1000
    g_2=12194**2*v**4/((v**2+20.6**2)*(v**2+12194**2)*((v**2+107.7**2)*(v**2+737.9**2))**0.5)
    g=20*math.log10(g_1)-20*math.log10(g_2)
    return g

def dB_p(VPa):
    p=VPa/a
    dBp=20*math.log10(p/0.00002)
    return dBp
        
FORMAT = pyaudio.paInt16    #[-32768 Ã· 32767]=2^16   
CHANNELS = 1                
RATE = 44100
CHUNK = 8192  #2^13 
dev_index = 0

p = pyaudio.PyAudio()
stream = p.open(
    format=FORMAT,
    channels=CHANNELS,
    rate=RATE,
    input=True,
    output=True,
    input_device_index = dev_index,
    frames_per_buffer=CHUNK)

#ÐºÐ¾Ñ€ÐµÐºÑ†Ñ–Ñ Ñ‡ÑƒÑ‚Ð»Ð¸Ð²Ð¾ÑÑ‚Ñ– Ð¼Ñ–ÐºÑ€Ð¾Ñ„Ð¾Ð½Ð°
#Ñ€Ð¾Ð·Ñ€Ð°Ñ…ÑƒÐ½Ð¾Ðº ÐºÐ¾ÐµÑ„Ñ–Ñ†Ñ–Ñ”Ð½Ñ‚Ð° Ð¿ÐµÑ€ÐµÑ‚Ð²Ð¾Ñ€ÐµÐ½Ð½Ñ

dBV = -40.0                #-47
dBV0 = dBV+(120-94)        #-21
print('dBV',dBV,dBV0)

a=np.power(10.0,dBV/20.0)   #4.47e-3
a0=np.power(10.0,dBV0/20.0) #89.13e-3
print('a',a,a0)

b0=np.power(2.0,15)        #32768
b=b0*a/a0                  #1642.29
print('b',b,b0)

c=a0/b0
print('c',c)

stream.start_stream()
data = np.frombuffer(stream.read(CHUNK),dtype=np.int16)
stream.stop_stream()
fps=RATE
Z=CHUNK

#Z=int(fps*T)
T=Z/fps
print('T',T)
#ÐºÑ€Ð¾Ðº Ñ‡Ð°ÑÑ‚Ð¾Ñ‚Ð¸ 
dv=1/T

dT=1/fps
# Ð²Ñ–ÑÑŒ Ñ‡Ð°ÑÑƒ
t=np.arange(Z)*dT
print('t', t[0], t[-1], len(t))

#Ð¿ÐµÑ€ÐµÑ‚Ð²Ð¾Ñ€ÐµÐ½Ð½Ñ Ð±Ñ–Ñ‚Ñ–Ð²
#data-->V/Pa
data = data*c

#Ð—Ð°Ð»ÐµÐ¶Ð½Ñ–ÑÑ‚ÑŒ Ð·Ð½Ð°Ñ‡ÐµÐ½ÑŒ (Ñƒ V/Pa) Ð²Ñ…Ñ–Ð´Ð½Ð¾Ð³Ð¾ ÑÐ¸Ð³Ð½Ð°Ð»Ñƒ Ð²Ñ–Ð´ Ñ‡Ð°ÑÑƒ
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_title(r'Ð“Ñ€Ð°Ñ„Ñ–Ðº Ð²Ñ…Ñ–Ð´Ð½Ð¾Ð³Ð¾ ÑÐ¸Ð³Ð½Ð°Ð»Ñƒ, V/Pa', fontsize=10, x=0.5, y=0)
ax.set_xlabel('Ð§Ð°Ñ, Ñ')
ax.set_ylabel('Ð—Ð½Ð°Ñ‡ÐµÐ½Ð½Ñ, V/Pa')
ax.plot(t[:1000:10], data[:1000:10], color='blue', marker='o', label="Ð“Ñ€Ð°Ñ„Ñ–Ðº Ð²Ñ…. ÑÐ¸Ð³Ð½Ð°Ð»Ñƒ data*c", linewidth=4)

#Ð”ÐŸÐ¤
y_1=fft(data)

#Ð²Ñ–ÑÑŒ Ñ‡Ð°ÑÑ‚Ð¾Ñ‚
L=len(y_1)
x = np.linspace(0, int((L)/T), L)
y1=[]
for i in range(L):
    y1.append(abs(y_1[i])/(L/2))

#Ð¿Ð¾Ð»Ð¾Ð²Ð¸Ð½Ð° ÑÐ¿ÐµÐºÑ‚Ñ€Ñƒ
Z1=int(L/2)

#ÐÐ§Ð¥
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_title(r'ÐÐ§Ð¥', fontsize=10, x=0.5, y=0)
ax.set_xlabel('Ð§Ð°ÑÑ‚Ð¾Ñ‚Ð°, Ð“Ñ†')
ax.set_ylabel('ÐÐ¼Ð¿Ð»Ñ–Ñ‚ÑƒÐ´Ð°, V/Pa')
ax.set_xscale('log')
ax.plot(x[:Z1], y1[:Z1], color='blue', marker='o', label="Ð“Ñ€Ð°Ñ„Ñ–Ðº ÐÐ§Ð¥", linewidth=4)
MaxVPa=max(y1)
Max_data=MaxVPa/c
print('Max_data=',Max_data)
V_maxA=x[np.argmax(y1)]
print('MaxVPa=',MaxVPa,'V/Pa')
print('V_maxA=',V_maxA,'Hz')
MaxPa=MaxVPa*1/a
print('MaxPa=',MaxPa,'Pa')
dB=20*math.log10(MaxVPa/a)
print('dB=',dB, 'dB')
MaxdBV=dB+dBV
print('MaxdBV=',MaxdBV,'dBV')
MaxdB=dB+94
print('MaxdB=',MaxdB,'dB')

y1=y1[:len(y1)//2]
L=len(y1)
x = np.linspace(0, int((L)/T), L)
dBA1=[]
dBA2=[]
dBA3=[]
Ps2,Ps3=0,0
for k in range(1,len(x)):
    dBA1.append(RA(x[k]))
    dBA2.append(dB_p(y1[k]))
    dBA3.append(dBA1[-1]+dBA2[-1])
    Ps2=Ps2+10**(dBA2[-1]/20)
    Ps3=Ps3+10**(dBA3[-1]/20)
dBAs2=20*math.log10(Ps2)
print('dBAs2=',dBAs2)
dBAs3=20*math.log10(Ps3)
print('dBAs3=',dBAs3)    
    
fig2 = plt.figure()
ax = fig2.add_subplot(1, 1, 1)
ax.set_xscale('log')
ax.plot(x[1:], dBA2, color='red', marker='.', label="Ð“Ñ€Ð°Ñ„Ñ–Ðº dBp", linewidth=1)
ax.plot(x[1:], dBA3, color='green', marker='.', label="Ð“Ñ€Ð°Ñ„Ñ–Ðº dBp3", linewidth=1)
s='MaxdB='+str(round(MaxdB,1))+' Ð´Ð‘; dBAs2='+str(round(dBAs2,1))+' Ð´Ð‘; dBAs3='+str(round(dBAs3,1))+' Ð´Ð‘Ð'
ax.set_title(s)
ax.set_xlabel('Ð§Ð°ÑÑ‚Ð¾Ñ‚Ð°, Ð“Ñ†')
ax.set_ylabel('Ð Ñ–Ð²ÐµÐ½ÑŒ Ñ‚Ð¸ÑÐºÑƒ, Ð´Ð‘')
plt.legend()
```
