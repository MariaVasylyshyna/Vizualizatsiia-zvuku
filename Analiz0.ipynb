```python
import numpy as np
from random import randint
import matplotlib.pyplot as plt
import math

# пошук ноти за її частотою з точністю e
def poshuk(vv, e):
    analiz=[]
    for i in range(len(muz)):
        if abs(vv-muz[i][0])<e:
            analiz.append([vv, muz[i]])
    return analiz
      
# створюємо масив випадкових амплітуд
N=300
A=[randint(0,400) for i in range(N)]

# створюємо масив частот
v1=250
v2=600
v = np.linspace(v1, v2, N)
# ширина суги частот
dv=v2-v1

# ставимо у відповідність номер амплітуди і частоту
# виводимо максимальну амплітуду і її номер (частоту)
V_max=v1+A.index(max(A))*dv/N

ss='A['+str(V_max)+']='+str(max(A))
print(ss)

# перелік нот в порядку збільшення її частоти
noty=['До','До#','Ре','Ре#','Мі','Фа','Фа#','Соль','Соль#','Ля','Ля#','Ci', 'До']
# коефіцієнт рівномірно-темперованої шкали; у скільки раз ноти відрізняються одна від одної
p=2**(1/12)

# створюємо масив частот і їх назв (назв нот)
muz=[]
for i in range(len(noty)):
    muz.append([440*(p**i)/(p**9), noty[i]])
    print(muz[i][0],'--',muz[i][1])
    
# масив різниць частот між нотами
muzD=[]
for i in range(1,len(muz)):
    muzD.append(muz[i][0]-muz[i-1][0])

%matplotlib inline

# figure - вікно, рисунок на якому будуть графіки
# axes - сам графік, система координат

fig = plt.figure()
# розкреслюємо рисунок на 3 частини
# вставляємо графік в першу комірку
ax1 = fig.add_subplot(1, 1, 1) 
ax1.grid()
ax1.set_title(r'АЧХ_'+ss, fontsize=10, x=0.5, y=1)

ax1.bar(v,A,width =1)

print('номер0-',A.index(max(A)), '--точність--',min(muzD))
print('------',poshuk(V_max, min(muzD)))


            
    
```
